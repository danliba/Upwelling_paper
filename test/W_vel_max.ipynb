```python
import xarray as xr
import matplotlib.pyplot as plt
from cmocean import cm # for oceanography-specific colormaps
```

```python
ds_w = xr.open_mfdataset('../../NHCS/hindcast/CROCO_BioEBUS_1990-2010/data/avg_Y*M*.nc',
                            chunks = {'time':1}).w.isel(s_rho=slice(15,None))
ds_w
```

```python
w = ds_w.where((ds_w.lat_rho.compute() >= -16) & (ds_w.lat_rho.compute() <= -5) & 
                  (ds_w.lon_rho.compute() >= -90) & (ds_w.lon_rho.compute() <= -65), drop=True)
w
```

```python
import pandas as pd
time_range = pd.date_range(start="1990-01-01", end="2010-12-01", freq="MS")  # 'MS' = Month Start
w = w.assign_coords(time=("time", time_range))
```
