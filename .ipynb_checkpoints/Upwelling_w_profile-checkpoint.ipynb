## Vertical velocity profile from 5S to 16S 

```python
import xarray as xr
import matplotlib.pyplot as plt
from cmocean import cm 
from scipy.io import loadmat
import numpy as np
import pandas as pd
import matplotlib.dates as mdates
```

```python
ds = xr.open_dataset('../../NHCS/hincast_1980-2015/croco_avg_Y1980M01.nc', 
                     chunks = {'time':1}).w
ds
```

```python
W = loadmat("../../NHCS/Processed/W_perfil_5_16.mat")
print(W.keys())
```

```python
time = pd.date_range(start='1980-01', end='2016-01', freq='M')
w_ds = xr.Dataset(
    data_vars={
        'w': (("time", "depth"), W['wnew'].transpose(1, 0)),
    },
    coords = {
        'depth':W['depths'][0,:],
        'time':time
}
)

w_ds
```

```python
# plot.contourf(x="lon", y="lat", col="time", col_wrap=5, vmin=-40,vmax=40,
#                                                                      cmap=cm.balance,norm=norm, extend='both',levels=levels)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,4))

w_ds.sel(depth=slice(0,-300)).w.plot(y='depth')
```

```python
w_ds.isel(time=0).sel(depth=slice(0,-300)).w.plot(y='depth')
w_ds.isel(time=10).sel(depth=slice(0,-300)).w.plot(y='depth')
w_ds.isel(time=20).sel(depth=slice(0,-300)).w.plot(y='depth')
plt.grid()
```

### Seasonality

```python
W_mmean = w_ds.groupby('time.month').mean(dim='time')
W_anomaly = w_ds.groupby('time.month') - W_mmean
W_mmean = W_mmean * 86400
```

```python
from matplotlib.colors import TwoSlopeNorm 

norm = TwoSlopeNorm(vmin=-1.5, vcenter=0, vmax=1.5)
levels=np.arange(-1.5,1.5+0.2,0.2)

W_mmean.sel(depth=slice(0,-200)).w.plot.contourf(y="depth",norm=norm, extend='both',levels=np.round(levels,1))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'];
_=plt.xticks(range(1,13), months)

plt.xlabel('Month')
plt.ylabel('Depth')
plt.title('W Vertical Velocity')

# plt.savefig('Figures/W_velocity_nhindcast'+'.png', format='png', dpi=300, transparent=False, bbox_inches='tight')
```

```python
W_mmean.sel(depth=slice(0,-200)).w.plot.line(y="depth")
plt.grid()
```

### TimeSeries

```python
W_anomaly = W_anomaly * 86400
```

```python
levels=np.arange(-3,3+0.4,0.4)
np.round(levels,1)
```

```python
norm = TwoSlopeNorm(vmin=-1.5, vcenter=0, vmax=1.5)
levels=np.arange(-1.5,1.5+0.2,0.2)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,4))

W_anomaly.sel(depth=slice(0,-300)).w.plot.contourf(y='depth',norm=norm, extend='both',levels=np.round(levels,1))
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator(base=2))  # Grid lines every 2 years
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format years as 'YYYY'

plt.title('W anomaly 5S - 16S')
plt.grid(linestyle='--', alpha=0.5)

plt.savefig('Figures/W_anomaly_5S_16S'+'.png', format='png', dpi=300, transparent=False, bbox_inches='tight')
```

```python
# norm = TwoSlopeNorm(vmin=-1.5, vcenter=0, vmax=1.5)
# levels=np.arange(-1.5,1.5+0.4,0.4)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,4))
W_anomaly.sel(depth=slice(0,-300)).sel(time=slice("1982-01","1984-07")).w.plot.contourf(y='depth',
                                                                                        norm=norm, extend='both',levels=np.round(levels,1))

plt.title('EN 1983-1984')
plt.grid(linestyle='--', alpha=0.5)

# plt.savefig('Figures/W_EN_1983_1984'+'.png', format='png', dpi=300, transparent=False, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,4))
W_anomaly.sel(depth=slice(0,-300)).sel(time=slice("1996-01","1999-01")).w.plot.contourf(y='depth',
                                                                                        norm=norm, extend='both',levels=np.round(levels,1))

plt.title('EN 1997-1998')
plt.grid(linestyle='--', alpha=0.5)

# plt.savefig('Figures/W_EN_1997_1998'+'.png', format='png', dpi=300, transparent=False, bbox_inches='tight')
```

```python
W_EN_8384 = W_anomaly.sel(depth=slice(0,-300)).sel(time=slice("1982-08","1983-12"))
W_EN_9798 = W_anomaly.sel(depth=slice(0,-300)).sel(time=slice("1997-04","1998-06"))

W_EN_8384
```

```python
# W_EN_8384.mean('time').w.plot(y='depth')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

w_mean_EN_8384 = W_EN_8384['w'].mean(dim='time')  # Mean across time for each depth
w_std_EN_8384 = W_EN_8384['w'].std(dim='time')    # Standard deviation across time for each depth

w_mean_EN_9798 = W_EN_9798['w'].mean(dim='time')  # Mean across time for each depth
w_std_EN_9798 = W_EN_9798['w'].std(dim='time')    # Standard deviation across time for each depth

axes[0].plot(w_mean_EN_8384, W_EN_8384['depth'], color='blue', label='Mean', linewidth=2)
axes[0].fill_betweenx(W_EN_8384['depth'], w_mean_EN_8384 - w_std_EN_8384, w_mean_EN_8384 + w_std_EN_8384, 
                  color='lightblue', alpha=0.5, label='±1 SD')
axes[0].set_title('EN 1982-08 to 1983-12')
axes[0].grid(linestyle='--', alpha=0.5)
axes[0].set_xlabel('W [m/day]')
axes[0].set_ylabel('m')
axes[0].set_xlim([-1, 1])
axes[0].axvline(0, color='black', linestyle='--', linewidth=1)

axes[1].plot(w_mean_EN_9798, W_EN_9798['depth'], color='green', label='Mean', linewidth=2)
axes[1].fill_betweenx(W_EN_9798['depth'], w_mean_EN_9798 - w_std_EN_9798, w_mean_EN_9798 + w_std_EN_9798, 
                  color='lightgreen', alpha=0.5, label='±1 SD')
axes[1].set_title('EN 1997-04 to 1998-06')
axes[1].grid(linestyle='--', alpha=0.5)
axes[1].set_xlabel('W [m/day]')
# axes[1].set_ylabel('m')
axes[1].set_xlim([-1, 1])
axes[1].axvline(0, color='black', linestyle='--', linewidth=1)

# plt.savefig('Figures/W_EN_std'+'.png', format='png', dpi=300, transparent=False, bbox_inches='tight')
```
